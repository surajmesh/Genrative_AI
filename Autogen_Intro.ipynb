```python
#Set API Endpoint
# Access the API key from the environment variable
config_list = [
        {
            'model': 'gpt-3.5-turbo',
            'api_key': os.environ.get("OPENAI_API_KEY")
        }
    ]
```

```python
# Configure the Language Model (LLM) with specified parameters
llm_config = {
        
        "seed": 42,  # seed for caching and reproducibility,
        "config_list": config_list, # a list of OpenAI API configurations
        "temperature": 0,  # temperature for sampling
        }
```

```python
# create an AssistantAgent named "assistant"
assistant = autogen.AssistantAgent(
    name="assistant",
    system_message = " You are a helpful AI assistant ",
    llm_config=llm_config, 
    is_termination_msg= lambda x: x.get("content", "").rstrip().endswith("TERMINATE") or x.get("content", "").strip() == "",
  
)
```

```python
# create a UserProxyAgent instance named "user_proxy"
user_proxy = autogen.UserProxyAgent(
                        name="user_proxy",
                        human_input_mode="TERMINATE",
                        max_consecutive_auto_reply=10,
                        is_termination_msg=lambda x: x.get("content", "").rstrip().endswith("TERMINATE") or x.get("content", "").strip() == "",
                        code_execution_config={"work_dir": "web"},
                        llm_config=llm_config, 
                        system_message = """" Reply TERMINATE if the task has been solved at full satisfactions,
                        Otherwise , reply CONTINUE , or the reason why the task is not solved yet. """ 
)

task = """ write code of fizz-buzz in python and store fizz_buzz.py file in web folder """
```

```python
# the assistant receives a message from the user_proxy, which contains the task description
user_proxy.initiate_chat(
    assistant,
    message=task,
)
```
